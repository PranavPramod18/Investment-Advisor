```python
import gspread
sa = gspread.service_account(filename="keys.json")
sheet = sa.open("Income _ Expense")
work_sheet = sheet.worksheet("inex")
```

```python
import pandas as pd
df = pd.DataFrame(work_sheet.get_all_values())
df
```

```python
df. columns=df. iloc[0]
df
```

```python
df = df.iloc[1: , :]
df
```
